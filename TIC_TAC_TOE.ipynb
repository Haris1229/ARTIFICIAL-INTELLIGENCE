```python
def print_board(board):
    print("\n" * 2)
    print("Tic Tac Toe 🎮\n")
    print(" 1 | 2 | 3 ")
    print("-----------")
    print(" 4 | 5 | 6 ")
    print("-----------")
    print(" 7 | 8 | 9 \n")
    print("Current board:\n")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} \n")

def check_winner(board):
    wins = [
        (0,1,2), (3,4,5), (6,7,8),
        (0,3,6), (1,4,7), (2,5,8),
        (0,4,8), (2,4,6)
    ]
    for a, b, c in wins:
        if board[a] != " " and board[a] == board[b] == board[c]:
            return board[a]
    return None

def is_full(board):
    return " " not in board

def get_move(board, player):
    while True:
        choice = input(f"Player {player}, enter a position (1-9), or 'quit': ").strip()
        if choice.lower() == "quit":
            return None
        if not choice.isdigit() or not (1 <= int(choice) <= 9):
            print("❌ Please enter a number between 1 and 9.")
            continue
        pos = int(choice) - 1
        if board[pos] != " ":
            print("❌ That spot is already taken. Choose another.")
            continue
        return pos

def play_tic_tac_toe():
    board = [" "] * 9
    current_player = "X"

    print("Welcome to Tic Tac Toe! Player X goes first. Type 'quit' anytime to exit.\n")
    print_board(board)

    while True:
        move = get_move(board, current_player)
        if move is None:
            print("Game ended by player.")
            return

        board[move] = current_player
        print_board(board)

        winner = check_winner(board)
        if winner:
            print(f"🎉 Player {winner} wins! Congratulations!")
            return

        if is_full(board):
            print("🤝 It's a draw!")
            return

        current_player = "O" if current_player == "X" else "X"

play_tic_tac_toe()
```
